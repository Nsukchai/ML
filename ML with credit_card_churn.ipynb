```python
def CalPercentage(feature,class_in_fea):  
    p = len(df_raw.loc[(df_raw[feature] == class_in_fea) & (df_raw['Attrition_Flag'] == 'Attrited Customer')])/len(df_raw.loc[(df_raw[feature] == class_in_fea)]) *100
    # pp = len(df_raw.loc[(df_raw[feature] == class_in_fea) & (df_raw['Attrition_Flag'] == 'Existing Customer')])/len(df_raw.loc[(df_raw[feature] == class_in_fea)])
    print('Percantages : ' + class_in_fea + ' in ' + feature + ' is Attrited Customer ' + str(round(p,3)) + ' %')
    print('Percantages : ' + class_in_fea + ' in ' + feature + ' is Existing Customer ' + str(round(abs(100 - p),3)) + ' %')
    # print(p,pp)
    return 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
df_r_model1 = df_raw.copy()

# df_r_model1['Attrition_Flag'] = df_r_model1['Attrition_Flag'].apply(lambda x: 0 if x=='Existing Customer' else 1)
y = df_r_model1.loc[:,'Attrition_Flag']
df_r_model1.drop(['Attrition_Flag'] ,axis=1, inplace=True)

x = pd.get_dummies(df_r_model1)
scale = StandardScaler()
x = scale.fit_transform(x)
```

```python
# x = df_r_model1.loc[:,df_r_model1.columns != 'Attrition_Flag']
# x = df_r_model1[x.columns]
# y = df_r_model1['Attrition_Flag']
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.1,random_state = 0)
```

Function

```python
from sklearn.model_selection import GridSearchCV

def GridSearch(model,params,x_train,y_train):
    gs = GridSearchCV(estimator = model,
                  param_grid = params,
                  cv = 5,
                  n_jobs = -1,
                  verbose = 2,
                  scoring='accuracy'
                 )
    gs.fit(x_train,y_train)
    return gs.best_params_
```

```python
def Report(y_t,y_p):
    
    print("Classification report \n=======================")
    print(classification_report(y_true=y_t, y_pred=y_p))
    print("Confusion matrix \n=======================")
    print(confusion_matrix(y_true=y_t, y_pred=y_p))

    return
```

ทำการตัด Avg_Open_To_Buy ด้วยเหตุผล 3 ข้อ
1. ข้อมูลระหว่าง Avg_Open_To_Buy กับ Credit_Limit มีค่าความสัมพันธ์กันที่สูงมาก สามารถเลือกแค่ Attribute เดียวเพื่อนำมาใช้งานได้
2. ข้อมูลระหว่าง Avg_Open_To_Buy กับ Credit_Limit เมื่อเทียบกับทุก Categorical Attributes แล้วกราฟมีลักษณะที่คล้ายกันมาก 
3. Avg_Open_To_Buy คือ ความต่างระหว่างวงเงินกับจำนวนเงินในบัตร และ Credit_Limit คือ วงเงินในบัตร ทำให้สมเหตุสมผลที่ยิ่งวงเงินสูงจำนวนเงินในบัตรเครดิตก็จะสูงตามได้ด้วย

```python
df_pre.drop(['Avg_Open_To_Buy'],axis=1,inplace=True)
```

ทำการตัด Total_Trans_Ct ด้วยเหตุผล 3 ข้อ 
1. ข้อมูลระหว่าง Total_Trans_Ct กับ Total_Trans_Amt มีค่าความสัมพันธ์กันที่สูง สามารถเลือกอันใดอันนึงได้
2. กราฟของทั้ง Total_Trans_Ct และ Total_Trans_Amt เมื่อเทียบกับทุก Categorical Attributes แล้วมีลักษณะที่คล้ายกัน คือ ค่ามัธยฐานของ Customer ที่ยกเลิกบริการจะน้อยกว่า Customer ที่ยังใช้บริการต่อ
3. Total_Trans_Amt มีข้อมูลส่วนใหญ่ที่อยู่มากกว่าค่ามัธยฐาน และความห่างของข้อมูลค่อนข้างมากจากค่ามัธยฐาน Total_Trans_Amt จะทำให้เห็นการกระจายตัวของข้อมูลได้ดีกว่า

```python
df_pre.drop(['Total_Trans_Ct'],axis=1,inplace=True)
```

ทำการตัด Avg_Utilization_Ratio ด้วยเหตุผล 3 ข้อ 
1. Total_Revolving_Bal กับ Avg_Utilization_Ratio มีลักษณะการกระจายตัวที่ใกล้เคียงกันเมื่อเทียบกับทุก Categorical Features คือ Customer จะเลิกบริการเมื่อ Total_Revolving_Bal หรือ Avg_Utilization_Ratio มีค่าน้อย
2. ข้อมูลระหว่าง Total_Revolving_Bal กับ Avg_Utilization_Ratio มีค่าความสัมพันธ์กันที่สูงขึ้นใน Customer ที่ยกเลิกเลิกบริการ
3. ข้อมูลส่วนใหญ่ของ Avg_Utilization_Ratio ค่าที่มากกว่าค่ามัธยฐานที่ค่อยข้างเยอะ
